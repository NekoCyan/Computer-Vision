```python
from PIL import Image
from imgTools import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
imageDir = "C:/Users/admin/Documents/GitHub/Computer-Vision/Images"
```

```python
imageList = get_image_list(imageDir)
image = imageList[0]
```

```python
plt.imshow(image)
```

```python
grayImage = image.convert("L")
plt.imshow(grayImage, cmap="gray")
```

```python
grayImageArray = np.array(grayImage)
print(grayImageArray)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(grayImageArray.flatten(), bins=128)
plt.title("Biểu đồ Histogram của ảnh")
plt.xlabel("Giá trị của pixel")
plt.ylabel("Số lượng")
plt.show()
```

```python
# Làm nổi bật đường biên
plt.figure()
plt.gray()
plt.contour(grayImageArray, origin="image")
plt.show()
```

```python
# Hiển thị ảnh trên trục tọa độ
plt.imshow(grayImageArray)

# Tạo một ình vẽ mới
plt.figure()

# Đặt màu sắc là xám
plt.gray()

# Vẽ biểu đồ đường biên (contour) của ảnh
plt.contour(grayImage, origin="image")

# Đảm bảo tỷ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis("equal")

# hiển thị ảnh
plt.show()
```
