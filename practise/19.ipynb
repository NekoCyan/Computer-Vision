```python
import cv2

def displayImages(title: str, waitKey: int, imgs: list):
    for i in range(len(imgs)):
        cv2.imshow(title + str(i), imgs[i])
    cv2.waitKey(waitKey)
    cv2.destroyAllWindows()
```

```python
path = "C:/Users/admin/Documents/GitHub/Computer-Vision/Images/animal.jpg"
img = cv2.imread(path)
x, y, z = img.shape
print(x, y, z)
```

```python
# Show image
cv2.imshow("Animal", img)
# Wait for a time
cv2.waitKey(10000)
# Close the window
cv2.destroyWindow("Animal")
```

```python
b, g, r = cv2.split(img)
cv2.imshow("Window", img)
cv2.imshow("Red", r)
cv2.imshow("Green", g)
cv2.imshow("Blue", b)
# Wait for a time
cv2.waitKey(0) # Press any key to close the window
# Close the window
cv2.destroyAllWindows()
```

```python
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.imshow("Window", img)
cv2.imshow("Gray", grayImage)
cv2.imshow("RGB", img2)
# Wait for a time
cv2.waitKey(30000)
# Close the window
cv2.destroyAllWindows()
```

```python
imgAlt = img.copy()

# Change pixel value
height, width, z = img.shape
quaterH, quaterW = height / 2, width / 2

# Green color
greenColor = (0, 255, 0)

# Change color of the image by one div four of the image (top left)
for y in range(int(quaterH)):
    for x in range(int(quaterW)):
        imgAlt[y, x] = greenColor

displayImages("Window", 30000, [imgAlt])
```

```python
imgAlt2 = img.copy()

# Change pixel value
height, width, z = img.shape
quaterH, quaterW = height / 2, width / 2

# Green color
greenColor = (0, 255, 0)

# Change color of the image by one div four of the image (bottom right)
for y in range(int(quaterH), int(height)):
    for x in range(int(quaterW), int(width)):
        imgAlt2[y, x] = greenColor

displayImages("Window", 30000, [imgAlt2])
```
